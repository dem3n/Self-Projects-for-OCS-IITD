```python
INPUT_DIR = '/content/drive/MyDrive/project/Results'
OUTPUT_DIR = '/content/drive/MyDrive/project/Cleaned'
```

```python
os.listdir(INPUT_DIR)
```

```python
for i in range(1, 10):
  exp_dir = f'{INPUT_DIR}/Experiment {i}'
  for results in os.listdir(exp_dir):
    df_path = os.path.join(exp_dir, results)
    # print(df_path)
    df = pd.read_csv(df_path)
    # print(df.columns)
    df = df[['Capacity', 'model_predict']]
    df.to_csv(os.path.join(OUTPUT_DIR, f'Experiment {i}', results), index=False)
```
